```python
from vllm import SamplingParams
```

```python
sampling_params = SamplingParams(max_tokens=4096,)
```

```python
from datasets import load_dataset
```

```python
dataset = load_dataset("Salesforce/xlam-function-calling-60k")
```

```python
import json
```

```python
def make_prompt(example):
    query = example['query']
    answers = json.loads(example['answers'])
    tools = json.loads(example['tools'])

    answer_names = [answer['name'] for answer in answers]
    selected_tools = [tool for tool in tools if tool['name'] in answer_names]
    
    prompt = f"""|im_start|>system
Given a user query and a list of JSON descriptions of possible function calls, evaluate each function in the list.
For each function, explain why it is a suitable choice for answering the query and provide a rationale for each argument value chosen.
Start your reasoning with: The correct function is ... 

Your output should have the following format:
[
    {{
        "function_name": "the name of the selected function",
        "reason": "the reasoning behind selecting the function and its arguments"
    }}
]
<|im_end|>

<|im_start|>user
Query: {query}
Function Calls: {answer}
Functions Descriptions: {selected_tools}
<|im_end|>

<|im_start|>assistant
    """
    
    prompt = prompt.strip()
    return {
        'prompt': prompt
    }
```

```python
prompt = make_prompt(dataset['train'][59956])
```

```python
result = llm.generate(prompt['prompt'], sampling_params)
```
