```python
dataset = load_dataset("0xayman/function_calling_dataset")
dataset
```

```python
from datasets import concatenate_datasets, DatasetDict, Dataset

# Access each split
train_data = dataset['train']
test_data = dataset['test']
dev_data = dataset['dev']

# Merge all splits into one
merged_dataset = concatenate_datasets([train_data, test_data, dev_data])
```

```python
DEFAULT_MAX_NEW_TOKENS = 200
```

```python
from vllm import SamplingParams
```

```python
def make_prompt(examples):
    query = examples['query']
    tools = examples['tools']
    prompt = f"""|im_start|>system
You are a helpful assistant with tool-calling capabilities.
Your sole responsibility is to output a valid JSON object.
Follow this exact format without deviation:
{{
"name": "<function_name>",
"arguments": {{ <argument_name>: <value>, ... }}
}}
Important instructions:
1. Output must be exclusively the JSON object, nothing else.
2. Do not add any explanations, context, or additional text.
3. No prefixes or suffixes of any kind—just the JSON.
4. Ensure there are no extra characters, whitespace, or line breaks outside the JSON.
5. Use only literal values for the arguments (numbers, strings, booleans).
6. There must be no quotes around the JSON object itself.

Example Output:
{{
  "name": "circle_area",  
  "arguments": {{
    "radius": 5  
  }}
}}

Available Tools:
{tools}
<|im_end|>\n<|im_start|>user\n{query}<|im_end|>
    """
    
    prompt = prompt.strip()
    return {
        'prompt': prompt,
    }
```

```python
dataset = merged_dataset.map(make_prompt)
```

```python
prompts = dataset['prompt']
```

```python
sampling_params = SamplingParams(temperature=1, top_p=0.99, top_k=9, min_tokens=2,max_tokens=200)
```

```python
result = llm.generate(dataset['prompt'], sampling_params)
```

```python
results = []
```

```python
for output in result:
    prompt = output.prompt
    generated_text = output.outputs[0].text.strip()
    results.append(generated_text)
```

```python
len(results)
```

```python
dataset = dataset.add_column('prediction', results)
```
