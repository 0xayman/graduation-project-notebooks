```python
dataset = load_dataset("Salesforce/xlam-function-calling-60k")
```

```python
sampling_params = SamplingParams(temperature=0.1, max_tokens=512)
```

```python
model_name = "Qwen/Qwen2.5-1.5B-Instruct"
```

```python
llm = LLM(model=model_name,
    gpu_memory_utilization=.95,
)
```

```python
import json
```

```python
def get_prompt_template(chat):
    query = chat['query']
    tools = json.loads(chat['tools'])
    answers = chat['answers']

    prompt = f"""<|im_start|>system
Cutting Knowledge Date: December 2023
Today Date: 14 Dec 2024

You have access to the following functions:
{tools}

Based on the query and the correct answer provided below, I want you to classify this function calling query to one of three categories.

# Categories:
1. Simple: the query can be fully answered by a single (only one) function call.
2. Parallel: the query can be answered by calling many (more than one) function. Function calls are independent of each other. this includes calling the same function many times with different arguments.
3. Chain: the query can be answered by calling many (more than one) dependent functions, the output from earlier functions is used as arguments to later functions.

Your answer should be ony ONE WORD indicating the category.
<|im_end|>

<|im_start|>user

Query: {query}
Correct answer: {answers}

<|im_end|>
"""


    prompt += "<|im_start|>assistant\nCategory:"

    return prompt
```

```python
def make_prompt(examples):
    prompt = get_prompt_template(examples)

    prompt = prompt.strip()
    return {
        'prompt': prompt,
    }
```

```python
dataset = dataset.map(make_prompt)
```
